### 极值处理方法


#### 分位数去极值

![title](./img/2.png)

```python
col = 'AAPL.O'
```

```python
def filter_extreme_percentile(series,min = 0.25,max = 0.75): #百分位法
    series = series.sort_values()
    q = series.quantile([min,max])
    return np.clip(series,q.iloc[0],q.iloc[1])
```

```python
percentile = filter_extreme_percentile(data[col])
```

```python
fig = plt.figure(figsize = (20, 8))
filter_extreme_percentile(data[col]).plot()
data[col].plot()
```

#### MAD又称为绝对值差中位数法

![title](./img/3.png)

- 数据集:(1, 1, 2, 2, 4, 6, 9)，它的中位数为2。
- 数据点到2的绝对偏差为(1, 1, 0, 0, 2, 4, 7)，该偏差列表的中位数为1（因为排序后的绝对偏差为(0, 0, 1, 1, 2, 4, 7)）
- 通常n都取1.4826，相对固定

```python
def filter_extreme_MAD(series,n): #MAD:中位数去极值
    median = series.quantile(0.5)
    new_median = ((series - median).abs()).quantile(0.5)
    max_range = median + n*new_median
    min_range = median - n*new_median
    return np.clip(series,min_range,max_range)
```

```python
mad = filter_extreme_MAD(data[col],1.4826)
```

```python
fig = plt.figure(figsize = (20, 8))
filter_extreme_MAD(data[col],1.4826).plot()
data[col].plot()
```

#### 3sigma方法

![title](./img/4.png)

```python
def filter_extreme_3sigma(series,n=3): 
    mean = series.mean()
    std = series.std()
    max_range = mean + n*std
    min_range = mean - n*std
    return np.clip(series,min_range,max_range)
```

```python
sigma = filter_extreme_3sigma(data[col],n=1)
```

```python
fig = plt.figure(figsize = (20, 8))
filter_extreme_3sigma(data[col],n=1).plot()
data[col].plot()
```

#### 中性化

简单概述就是要 **提纯**

在量化交易中，我们会经常使用某种指标或者多种指标来对股票池进行筛选，这些用于选股的指标一般被称为因子。在使用这些因子进行选股时，有时会因为其它因子的影响，而导致选出来的股票具有一些我们不希望看到的偏向。

例如，市净率pb会与市值有很高的相关性，这时如果我们使用未进行市值中性化的市净率，选股的结果会比较集中。

（市净率（Price-to-Book Ratio，简称P/B PBR）指的是每股股价与每股净资产的比率。 市净率可用于股票投资分析，一般来说市净率较低的股票，投资价值较高，相反，则投资价值较低）
